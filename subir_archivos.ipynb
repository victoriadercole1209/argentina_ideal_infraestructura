```python
#Creamos un bucket
storage_client = storage.Client()
bucket_name = 'argentina_ideal_grupo5_2025'
bucket = storage_client.create_bucket(bucket_name)
```

```python
carpetas = ['Archivos_Maestro', 'Archivos_Stock', 'Archivos_VentaClientes']
base_path = '/content/drive/MyDrive/INFRA-TP-FINAL/generated_data/'

# Recorrer carpetas y subcarpetas
for carpeta in carpetas:
    carpeta_path = os.path.join(base_path, carpeta)
    for root, dirs, files in os.walk(carpeta_path):
        for archivo in files:
            ruta_local = os.path.join(root, archivo)
            # Generar la ruta relativa para mantener la estructura
            ruta_relativa = os.path.relpath(ruta_local, base_path)
            blob = bucket.blob(ruta_relativa)
            blob.upload_from_filename(ruta_local)
            print(f'Subido: {ruta_relativa}')
```
